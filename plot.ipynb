```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_3 = pd.read_csv('frame_50_3.csv')
y1 = result_3["time"].tolist() 
y2 = result_3["the"].tolist() 
y3 = result_3["offset"].tolist()
x= result_3["frame"].tolist() 
```

```python
plt.figure(figsize=(15,10))
plt.plot(x, y1, label = "QUIC")
plt.plot(x, y2, label = "Theoretical")
plt.plot(x, y3, label = "Offset")
#plt.ylim(-0.0, 0.05)
#plt.xlim(1, 200)
plt.xlabel('Frame ID',fontsize=20)
plt.ylabel('Frame latency in seconds',fontsize=20)
plt.title('Per Frame Latency, Framesize=50KB ',fontsize=35)
plt.grid(True,color='black', linestyle='-', linewidth=0.25)
plt.legend(fontsize=20,loc="upper right")
plt.show()
```
